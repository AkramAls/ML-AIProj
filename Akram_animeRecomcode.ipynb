```python
anime_data = anime_review.copy()
counts = anime_data["anime_uid"].value_counts() # use to create a pivot table
anime_data = anime_data[anime_data["anime_uid"].isin(counts[counts >= 50].index)] 
#limit of 50 rating per user as this help even things out
```

```python
anime_pivot_temp = anime_data.pivot_table(index="title",columns="user_id",values="user score").fillna(0)
anime_pivot_temp.head(50)
```

```python
import re # some anime have titles that effect the outcome of the program or prevent it from running at all
def text_cleaning(text):
    text = re.sub(r'&quot;', '', text)
    text = re.sub(r'.hack//', '', text)
    text = re.sub(r'&#039;', '', text)
    text = re.sub(r'A&#039;s', '', text)
    text = re.sub(r'I&#039;', 'I\'', text)
    text = re.sub(r'&amp;', 'and', text)
    text = re.sub(r'★', '', text)
    return text

anime_data["title"] = anime_data["title"].apply(text_cleaning)
```

```python
anime_pivot = anime_data.pivot_table(index="title",columns="user_id",values="user score").fillna(0)
print("After Cleaning the animes names, let's see how it looks like.")
anime_pivot.head(50)
```

```python
from scipy.sparse import csr_matrix
#Creating a sparse matrix
anime_matrix = csr_matrix(anime_pivot.values)

from sklearn.neighbors import NearestNeighbors

#Fitting the model
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(anime_matrix)
```

```python
query_no = np.random.choice(anime_pivot.shape[0]) # random anime title and finding recommendation
print(f"We found a recommendation for {query_no} one which is {anime_pivot.index[query_no]}.")
distances, indices = model_knn.kneighbors(anime_pivot.iloc[query_no,:].values.reshape(1, -1), n_neighbors = 6)
# KNN is 6 so we can output 5 recommendations 
```

```python
no = []
title = []
distance = []
score = []

for i in range(0, len(distances.flatten())):
    if i == 0:
        print(f"Recommendations for {anime_pivot.index[query_no]} viewers :\n")
    else:
        #  print(f"{i}: {anime_pivot.index[indices.flatten()[i]]} , with distance of {distances.flatten()[i]}")        
        no.append(i)
        title.append(anime_pivot.index[indices.flatten()[i]])
        distance.append(distances.flatten()[i])
        score.append(*df_anime[df_anime["title"]==anime_pivot.index[indices.flatten()[i]]]['overall score'].values)

dic = {"No" : no, "Anime Title" : title, "Score" : score}
recommendation = pd.DataFrame(data = dic)
recommendation.set_index("No", inplace = True)
recommendation.style.set_properties(**{"background-color": "black","color":"white","border": "1.5px  solid black"})
```
